## 1. BankNote Authentication Veri Seti

Banknot görüntülerinden çıkarılan istatistiksel özellikler ile sahte/gerçek banknot sınıflandırması.

- **Özellikler**: variance, skewness, curtosis, entropy
- **Hedef**: class (0: gerçek, 1: sahte)
- **Toplam örnek**: 1372

```python
# Veri setini yükle
loader_bn = DataLoader(random_state=42)
df_bn = loader_bn.load_banknote('../data/BankNote_Authentication.csv')
df_bn.head(10)
```

```python
# Keşifsel Veri Analizi
loader_bn.explore()
```

```python
# EDA Grafikleri
loader_bn.plot_eda(save_dir='../results/banknote')
```

```python
# Veri Bölme ve Standardizasyon
X_train_bn, X_val_bn, X_test_bn, y_train_bn, y_val_bn, y_test_bn = loader_bn.split_data(
    test_size=0.15, val_size=0.15, scaling='standard'
)
```

```python
# Standardizasyon sonrası istatistikler
print('\nStandardizasyon sonrası Train seti istatistikleri:')
print(f'  Ortalama: {X_train_bn.mean(axis=0).round(4)}')
print(f'  Std     : {X_train_bn.std(axis=0).round(4)}')
print(f'\n  X_train shape: {X_train_bn.shape}')
print(f'  X_val   shape: {X_val_bn.shape}')
print(f'  X_test  shape: {X_test_bn.shape}')
print(f'  y_train shape: {y_train_bn.shape}')
print(f'  y_val   shape: {y_val_bn.shape}')
print(f'  y_test  shape: {y_test_bn.shape}')
```

## 3. Pairplot - Özellik İlişkileri

Her iki veri seti için sınıflara göre özellik çiftleri arasındaki ilişkilerin görselleştirilmesi.

```python
os.makedirs('../results/banknote', exist_ok=True)
# BankNote Pairplot
fig = sns.pairplot(df_bn, hue='class', palette='Set1', diag_kind='kde', corner=True)
fig.figure.suptitle('BankNote - Özellik Çift Grafikleri', y=1.02, fontsize=14, fontweight='bold')
plt.savefig('../results/banknote/BankNote_pairplot.png', dpi=150, bbox_inches='tight')
plt.show()
```
